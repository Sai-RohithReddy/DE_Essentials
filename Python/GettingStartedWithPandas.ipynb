```python
import json
import pandas as pd
```

```python
def get_column_names(schemas, ds_name, sorting_key='column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    return [col['column_name'] for col in columns]
```

```python
schema = json.load(open('data/retail_db/schemas.json'))
orders_columns = get_column_names(schema, 'orders')
```

```python
orders = pd.read_csv('data/retail_db/part-00000', names=orders_columns)
```

```python
customers_columns = get_column_names(schema, 'customers')
```

```python
customers = pd.read_csv('data/retail_db/part-00000', names=customers_columns)
```

```python
orders.head()
```

```python
customers.head()
```

```python
# Performing Inner Join
```

```python
customers = customers.set_index('customer_id')
orders = orders.set_index('order_customer_id')
```

```python
customers.head()
```

```python
orders.head()
```

```python
customer_orders = customers. \
    join(orders, how='inner')
```

```python
customer_orders.head()
```

```python
customer_orders.shape
```

```python
customer_orders. \
    reset_index(names='customer_id'). \
    groupby('customer_id')['customer_id']. \
    agg(order_count='count'). \
    reset_index(). \
    query('order_count >= 10')
```
